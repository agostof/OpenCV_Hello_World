# OpenCV Hello World (Voilà  demo)

This demo runs the "OpenCV Hello World" Notebook on an uploaded image by the user using the [Voilà](https://github.com/voila-dashboards/voila) extension.

The resulting view will only display the markdown content and plots.


Steps:
1. Please upload an image from your computer. May I suggest some apples: [apples-on-wooden-table-w1024.png](data/apples-on-wooden-table-w1024.png)
2. Load image.
3. Click the "Run Analysis Link" when ready. This will appear below the image.


```python
# Function to handle button click
VALUES = None
from PIL import Image
import io
import numpy as np
import cv2
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import json

global content
# Create a file upload widget
uploader = widgets.FileUpload(
    accept='',  # Specify file types, e.g., '.txt', '.pdf', etc.
    multiple=False,  # Allow multiple files to be selected
    #description= ""
)

# Create a button widget
button = widgets.Button(
    description='Load Image'
)

# Create a button widget
open_notebook = widgets.Button(
    description='Configure Analysis Notebook'
)

output = widgets.Output()

def process_image(image_file):
    image = Image.open(image_file)
    return image


def run_analysis(b):
    with output:
        display(HTML('<a href="OpenCV_Hello_Dashboard.ipynb">Run Analysis (New Window or Tab)</a>'))

def on_button_clicked(b):
    global VALUES
    # Clear previous output
    #clear_output()
    with output:
        output.clear_output()

        #clear_output()
        if uploader.value:
            # Get the uploaded file; uploader.value is a dictionary
            VALUES = uploader.value
            
            #for uploaded_file in uploader.value:
            uploaded_file = next(iter(uploader.value))
            if uploaded_file:
                image_file = io.BytesIO(uploaded_file['content'])
                metadata = {}
                metadata['original_name'] = uploaded_file['name']
                metadata['stored_named'] = 'data/uploaded_input.jpg'
                json.dump(metadata, open('data/metadata.json', "w"))
                open(metadata['stored_named'], "wb").write(uploaded_file['content'])
                
                # Show the uploaded and processed images
                image_data = process_image(image_file)
                display(image_data)
                #display(open_notebook)
                display(HTML('<a href="OpenCV_Hello_Dashboard.ipynb">Run Analysis (New Window or Tab)</a>'))
                
            else:
                print("No file uploaded.")
        else:
            print("Upload a file to process.")
```

```python
# Use an alternate picture for the demo:
# apples-on-wooden-table-w1024


# Bind the button click event
button.on_click(on_button_clicked)
open_notebook.on_click(run_analysis)

# Display widgets
display(uploader, button, output)
```
